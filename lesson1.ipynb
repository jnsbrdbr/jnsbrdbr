```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
fig,axes=plt.subplots(3,1,figsize=(8,20))
k_neighbors=[1,3,5]
x=np.linspace(-2.9,2.9,500).reshape(-1,1)
# train_acc=[]
# test_acc=[]

for k,ax in zip(k_neighbors,axes):
  knn= KNeighborsRegressor(n_neighbors=k)
  knn.fit(X_train, y_train)
  ax.plot(x,knn.predict(x))
  ax.plot(X_train,y_train,'s',c='green')
  ax.plot(X_test,y_test,'s',c='black')
```
